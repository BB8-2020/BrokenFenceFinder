```python
## Importing Necessary Modules
import requests # to get image from the web
import shutil # to save it locally
import json
import os
import cv2
from imageio import imread
import numpy as np
import time
import tensorflow as tf
```

```python

try:
    # creating a folder named data
    if not os.path.exists('images'):
        os.makedirs('images')
    if not os.path.exists('masks'):
        os.makedirs('masks')
        

# if not created then raise error
except OSError:
    print('Error: Creating directory of images')


with open('labels.json') as json_file:
    data = json.load(json_file)
    c = 0
    for row in data:
        if c == 300:
            break
        if c % 50 == 0:
            print(c)
        c += 1
        filename = row['ID'] + '.png'
        image_url = row['Labeled Data']
        try:
            mask_url = row['Label']['objects'][0]['instanceURI']
        except KeyError:
            mask_url = False
            from PIL import Image
            img = Image.new("RGB", (1920, 1080), (0, 0, 0))
            img.save("masks/" + filename, "PNG")
            print("did sometin else")
            
            
        if mask_url:
#             print(mask_url)
            time.sleep(.2)
            mask_img = imread(mask_url)
            mask_img = mask_img // 255
            mask_img = np.delete(mask_img, np.s_[:3], 2)
            mask_img = np.reshape(mask_img, (1080, 1920))
#             cv2.resize(mask_img, dsize=(1056, 140), interpolation=cv2.INTER_CUBIC)
            cv2.imwrite('masks/' + filename, mask_img)
            
            
        
        # Open the url image, set stream to True, this will return the stream content.
        r = requests.get(image_url, stream = True)

        # Check if the image was retrieved successfully
        if r.status_code == 200:
            # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
            r.raw.decode_content = True

            # Open a local file with wb ( write binary ) permission.
            with open('images/' + filename,'wb') as f:
                shutil.copyfileobj(r.raw, f)

#             print('Image sucessfully Downloaded: ',filename)
        else:
            print('Image Couldn\'t be retreived')
            
        
#         if mask_url:
#             r = requests.get(mask_url, stream = True)

#             # Check if the image was retrieved successfully
#             if r.status_code == 200:
#                 # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
#                 r.raw.decode_content = True

#                 # Open a local file with wb ( write binary ) permission.
#                 with open('masks/' + filename,'wb') as f:
#                     shutil.copyfileobj(r.raw, f)

#                 print('Image sucessfully Downloaded: ',filename)
#             else:
#                 print('Image Couldn\'t be retreived')
```

```python
from keras_segmentation.models.unet import vgg_unet
```

```python
model = vgg_unet(n_classes=2, input_height=480, input_width=832  )
```

```python
model.train(
    train_images =  "images/",
    train_annotations = "masks/", epochs=35
)
```

```python
out = model.predict_segmentation(
    inp="test3.png",
    out_fname="out.png"
)
```
