```python
import numpy as np
import cv2
```

```python
path = '../media/'
image = cv2.imread(path+'fence.png')
```

# Method 1:
Histogram equalizer

```python
imggray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
eq = cv2.equalizeHist(imggray)
cv2.imwrite(path+'equalhist.png',eq)
```

# Method 2:
CLAHE (Contrast Limited Adaptive Histogram Equalization)

```python
clahe = cv2.createCLAHE(clipLimit=0.4, tileGridSize=(64,64))
cl1 = clahe.apply(imggray)

cv2.imwrite(path+'clahe.png',cl1)
```

# Testing:
finding edges in contrastboosted photos.

# Threshold and contours.

```python
ret, thresh = cv2.threshold(eq, 75, 255, 0)

cv2.imwrite(path+'thresh_equalhist.png', thresh)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
img = cv2.drawContours(image, contours, -1, (0,255,0), 2)

cv2.imwrite(path+'contours_equalhist.png', img)
```

```python
ret, thresh = cv2.threshold(cl1, 95, 255, 0)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
img = cv2.drawContours(image, contours, -1, (0,255,0), 2)

cv2.imwrite(path+'contours_clahe.png', img)
```

# Bilateral filter

```python
image = cv2.imread(path+'fence.png')

imggray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
eq = cv2.equalizeHist(imggray)
cv2.imwrite(path+'equalhist.png',eq)

sharp = cv2.bilateralFilter(eq, 5, 35, 35)
cv2.imwrite(path+'sharp_equalhist.png', sharp)
edged = cv2.Canny(sharp, 200, 500)
cv2.imwrite(path+'bilateralf_equalhist.png', edged)
```

```python
image = cv2.imread(path+'fence.png')

clahe = cv2.createCLAHE(clipLimit=0.4, tileGridSize=(64,64))
cl1 = clahe.apply(imggray)

sharp = cv2.bilateralFilter(cl1, 5, 17, 17)
edged = cv2.Canny(sharp, 30, 200)
cv2.imwrite(path+'bilateralf_clahe.png', edged)
```
