```python
import numpy as np
import cv2
import os
```

```python
try:
    if not os.path.exists('filtered'):
        os.makedirs('filtered')
        print("filtered dir created!")
except OSError:
    print('Error: Creating directory of filtered')
```

```python
rootdir = os.path.abspath(os.curdir) + '\images'
rootdir
```

```python
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        filepath = subdir + os.sep + file
        filename = filepath.split('\\')[-1]
        try:
                image = cv2.imread(filepath)
                imggray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                eq = cv2.equalizeHist(imggray)
                ret, thresh = cv2.threshold(eq, 120, 255, 0)
                cv2.imwrite('filtered/' + filename, thresh)
        except:
            print("something went wrong...")
```
